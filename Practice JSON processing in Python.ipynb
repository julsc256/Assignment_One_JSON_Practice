## Question 3: Make a histogram of the frequencies of earthquakes in each city. Did any have more than one?

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
for i in range(0,14):
    print(data['features'][i]['properties']['place'][9:29])
```

```python
places = [data['features'][i]['properties']['place'][9:29] for i in range(0,14)]
```

```python
plt.figure(figsize=(30,10))
plt.hist(places)
plt.title('Frequency of Earthquakes')
plt.xlabel('Cities')
plt.ylabel('Frequency');
```

### Answer:
The Geysers, CA is the city that shows to have the highest frequency of 3 earthquakes.
